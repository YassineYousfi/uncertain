```python
import numpy as np

from sklearn.gaussian_process import GaussianProcessRegressor
import matplotlib.pyplot as plt 

from data import squiggle
```

```python
x = np.linspace(-10,10,1000)
plt.plot(x, squiggle(x))
plt.show()
```

```python
x = np.random.uniform(low=-4, high=4, size=(1000,1))
y = squiggle(x)
gpr = GaussianProcessRegressor( random_state=0).fit(x, y)
```

```python
x_test = np.linspace(-10,10,200)[..., None]
mu, sigma = gpr.predict(x_test, return_std=True)
plt.plot(x_test, mu)
plt.fill_between(
    x_test.ravel(),
    mu.ravel() - sigma.ravel(),
    mu.ravel() + sigma.ravel(),
    alpha=0.5,
)
plt.show()
```
